## SICP 习题 （3.13）解题总结: 吞食自己的蛇

SICP 习题 3.12 为我们展示了一个非常有趣的数据结构，主要是为了让我们更直观地了解set-cdr!函数。

题目给我们展示了这样一个函数：

```scheme
(define (make-cycle x)
  (set-cdr! (last-pair x) x)
  x)
```

这个函数里面调用的(last-pair)函数是习题3.11里提到的函数，它可以获得一个列表的最后一个元素。这里稍微复习一下，一个列表的最后一个元素的cdr本来是指向()的。现在我们通过(make-cycle)函数把x最后一个元素的cdr指向了x自己。这就意味着这是一个环链，尾巴连着头！


关于(last-pair)和列表最后一个元素的cdr值。

我们通过下面的代码重温一下：

```scheme
(define (last-pair x)
  (if (null? (cdr x))
      x
      (last-pair (cdr x))))

(define a (last-pair '(1 2 3 4)))
```

```scheme
a
```

```scheme
(cdr a)
```

题目问我们，如果我们定义了一下变量：

    (define z (make-cycle (list 'a 'b 'c)))
    
那么求`(last-pair z)`会发生什么？

既然是一个头咬着尾的列表，那么它就没有last-pair，这段代码会不断递归，无法退出。不信的话运行一下这段代码就可以了：

```scheme
(define z (make-cycle (list 'a 'b 'c)))

```

```scheme
(last-pair z)
```
