## SICP 习题 （2.64）解题总结：排序列表到平衡二叉树的转换

SICP 习题2.64要我们考察一个把排序列表转换成平衡二叉树的函数，要我们分析输出结果和函数运行的复杂度。

代码书里面直接给出了，所以一股脑先把代码撸下来，加上相关的辅助函数，代码如下：

```scheme
(define (list->tree elements)
  (car (partial-tree elements (length elements))))

(define (is-tree? tree)
  (if (pair? tree)
      (if (= (length tree) 3)
	  #t
	  #f)
      #f))

(define (entry tree) 
  (if (is-tree? tree) 
       (car tree)
       tree))


(define (left-branch tree)
  (if (is-tree? tree)
      (cadr tree)
      '()))

(define (right-branch tree) 
  (if (is-tree? tree)
      (caddr tree)
      '()))

(define (make-tree entry left right)
  (list entry left right))


(define (partial-tree elts n)
  (if (= n 0)
      (cons '() elts)
      (let ((left-size (quotient (- n 1) 2)))
	(let ((left-result (partial-tree elts left-size)))
	  (let ((left-tree (car left-result))
		(non-left-elts (cdr left-result))
		(right-size (- n (+ left-size 1))))
	    (let ((this-entry (car non-left-elts))
		  (right-result (partial-tree (cdr non-left-elts)
					    right-size)))
	      (let ((right-tree (car right-result))
		    (remaining-elts (cdr right-result)))
		(cons (make-tree this-entry left-tree right-tree)
		      remaining-elts))))))))


(define (partial-tree-with-detail elts n)
  (display "------------------------------") (newline)
  (display "handling: ") (display elts) (newline)
  (display "with n: ") (display n) (newline)
  (if (= n 0)
      (cons '() elts)
      (let ((left-size (quotient (- n 1) 2)))
	(display "left-size is: ") (display left-size) (newline)
	(let ((left-result (partial-tree-with-detail elts left-size)))
	  (display "left-result is: ") (display left-result) (newline)
	  (let ((left-tree (car left-result))
		(non-left-elts (cdr left-result))
		(right-size (- n (+ left-size 1))))
	    (display "left-tree is: ") (display left-tree) (newline)
	    (display "non-left-elts is: ") (display non-left-elts) (newline)
	    (display "right-size is: ") (display right-size) (newline)
	    (let ((this-entry (car non-left-elts))
		  (right-result (partial-tree-with-detail (cdr non-left-elts)
					    right-size)))
	      (display "this-entry is: ") (display this-entry) (newline)
	      (display "right-result is: ") (display right-result) (newline)
	      (let ((right-tree (car right-result))
		    (remaining-elts (cdr right-result)))
		(cons (make-tree this-entry left-tree right-tree)
		      remaining-elts))))))))

(define (start-test-2-64)
  (define test-elements '(1 3 4 5 6 7 9 10))
  (display "the result is:------------") (newline)
  (display (car (partial-tree-with-detail test-elements (length test-elements)))) (newline)
  )
```

测试一番，我在partial-tree-with-detail函数里加上了很多分析输出，便于分析程序执行过程：

```scheme
(start-test-2-64)
```

虽然输出很复杂，但是这个函数的复杂度是O(n),过程就是不断折半，然后分别处理左树和右树。

不断折半，听起来像个O(ln(n))的复杂度，其实只是折出来的树的高度是ln(n), 但是折半后还需要对每个元素进行操作,这种需要对叶子结点做处理的函数复杂度不能逃过O(n).

另外，题目2.65也透露了一点信息，2.65要求我们用2.65这个函数构建 union-set和intersection-set函数，要求复杂度是O(n)。
